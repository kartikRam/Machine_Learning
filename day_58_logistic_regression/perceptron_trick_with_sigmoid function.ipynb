```python
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
import numpy as np
import matplotlib.pyplot as plt
```

```python
X,y= make_classification(n_samples=100,n_classes=2,n_features=2,n_informative=1,n_redundant=0,n_clusters_per_class=1,random_state=41,hypercube=False, class_sep=30)
```

```python
X.shape
#y.shape
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1], c=y, cmap='winter',s=100)
plt.show()
```

## Logistic Regression SkLearn Class

```python
lor=LogisticRegression()
lor.fit(X,y)
```

```python
m1 = -(lor.coef_[0][0]/lor.coef_[0][1])
b1 = -(lor.intercept_/lor.coef_[0][1])
x_input1 = np.linspace(-3,3,100)
y_input1 = m1*x_input1 + b1
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1], c=y, cmap='winter',s=100)
plt.plot(x_input1,y_input1,c='black',linewidth=3)
plt.ylim(-3,3)
plt.show()
```

## Perceptron Trick uisng Step Function

```python
def step(num):
    return 1 if num>0 else 0
```

```python
def Perceptron(X,y):
    X = np.insert(X,0,1,axis=1)
    lr = 0.1
    weights = np.ones(X.shape[1])
    size = X.shape[0] -1
    for i in range(200):
        
        j = np.random.randint(0,size)
        if j == 100:
            print(j)
        y_hat = step(np.dot(X[j],weights))
        weights = weights + lr * (y[j] - y_hat) * X[j]
        
    return weights[0],weights[1:]
```

```python
intercept , coeff = Perceptron(X,y)
m2 = - coeff[0]/coeff[1]
b2 = - intercept / coeff[1]
```

```python
x_input2 = np.linspace(-3,3,100)
y_input2 = m2*x_input2 + b2
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1], c=y, cmap='winter',s=100)
plt.plot(x_input1,y_input1,c='black',linewidth=3)
plt.plot(x_input2,y_input2,c="r",linewidth=3)
plt.ylim(-3,2)
plt.show()
```

## Perceptron Trick with Sigmoid Function

```python
def sigmoid(num):
    return 1 / (1 + np.exp(-num))
```

```python
def Perceptron(X,y):
    X = np.insert(X,0,1,axis=1)
    lr = 0.1
    weights = np.ones(X.shape[1])
    size = X.shape[0] -1
    for i in range(200):
        
        j = np.random.randint(0,size)
        if j == 100:
            print(j)
        y_hat = sigmoid(np.dot(X[j],weights))
        weights = weights + lr * (y[j] - y_hat) * X[j]
        
    return weights[0],weights[1:]
```

```python
intercept_,coef_ = Perceptron(X,y)
```

```python
m3 = -(coef_[0]/coef_[1])
b3 = -(intercept_/coef_[1])
```

```python
y_input3 = m3*x_input2 + b3
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1], c=y, cmap='winter',s=100)
plt.plot(x_input1,y_input1,c='black',linewidth=3)
plt.plot(x_input2,y_input2,c="r",linewidth=3)
plt.plot(x_input2,y_input3,c='brown',linewidth=3)
plt.ylim(-3,2)
plt.show()
```
