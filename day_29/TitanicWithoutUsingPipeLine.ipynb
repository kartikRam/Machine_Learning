```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
```

```python
df=pd.read_csv('../train.csv')
```

```python
df.head()
```

```python
df=df.drop(columns=['PassengerId','Ticket','Name','Cabin'])
```

```python
df.head()
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(df.drop(columns=['Survived'],axis=1),df['Survived'],test_size=0.2,random_state=0)
```

```python
print(X_train.head(2))
print(X_test.head(2))
print(X_test['Embarked'].value_counts())
print(X_train['Embarked'].value_counts())
```

```python
Y_train.head()
```

```python
df.isnull().sum()
```

```python
#Applying Imputaion

si_age=SimpleImputer()
si_embarked=SimpleImputer(strategy='most_frequent')

X_train_age=si_age.fit_transform(X_train[['Age']])
X_test_age=si_age.fit_transform(X_test[['Age']])

X_train_embarked=si_embarked.fit_transform(X_train[['Embarked']])
X_test_embarked=si_embarked.fit_transform(X_test[['Embarked']])

X_train_embarked=pd.DataFrame(X_train_embarked)
X_test_embarked=pd.DataFrame(X_test_embarked)

X_train_embarked.isnull().sum()
```

```python
# one Hot encoding on sex and Embarked
ohe_sex=OneHotEncoder(sparse_output=False, drop='first')
ohe_embarked=OneHotEncoder(sparse_output=False, drop='first')

X_train_sex=ohe_sex.fit_transform(X_train[['Sex']])
X_test_sex=ohe_sex.fit_transform(X_test[['Sex']])

X_train_embarked=ohe_embarked.fit_transform(X_train_embarked)
X_test_embarked=ohe_embarked.fit_transform(X_test_embarked)
```

```python
X_train_sex=pd.DataFrame(X_train_sex)
X_test_sex=pd.DataFrame(X_test_sex)
#X_train_sex.column=['sex']
X_train_embarked=pd.DataFrame(X_train_embarked)
X_test_embarked=pd.DataFrame(X_test_embarked)


print("X_train_Sex\n",X_train_sex.sample(5))
print("X_test_Sex\n",X_test_sex.sample(5))


print("X_train_Embarked\n",X_train_embarked.sample(5))
print("X_test_Embarked\n",X_test_embarked.sample(5))
```

```python
X_test.head()
```

```python
X_train_rem=X_train.drop(columns=['Sex','Age','Embarked'])
X_test_rem=X_test.drop(columns=['Sex','Age','Embarked'])
```

```python
X_test
```

```python
X_train_transformed = np.concatenate((X_train_rem,X_train_age,X_train_embarked,X_train_sex), axis=1)
X_train_transformed = pd.DataFrame(X_train_transformed)

X_test_transformed = np.concatenate((X_test_rem,X_test_age,X_test_embarked,X_test_sex), axis=1)
X_test_transformed = pd.DataFrame(X_test_transformed)
```

```python
X_train_transformed.sample(5)
```

```python
clf = DecisionTreeClassifier()
clf.fit(X_train_transformed,Y_train)
```

```python
y_pred = clf.predict(X_test_transformed)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,y_pred)
```

```python
import pickle
pickle.dump(ohe_sex,open('models/ohe_sex.pkl','wb'))
pickle.dump(ohe_embarked,open('models/ohe_embarked.pkl','wb'))
pickle.dump(clf,open('models/clf.pkl','wb'))
```
