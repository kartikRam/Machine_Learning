```python
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt
```

```python
X,y = make_regression(n_samples=100, n_features=1,n_informative=1,n_targets=1,noise=15,random_state=13)
```

```python
plt.scatter(X,y)
```

## Alpha = 0  

```python
## 1st line where alpha = 0
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X,y)
print(lr.coef_)
print(lr.intercept_)
```

## Alpha = 10

```python
from sklearn.linear_model import Ridge
rgd1 = Ridge(alpha=10)
rgd1.fit(X,y)
print(rgd1.coef_)
print(rgd1.intercept_)
```

## Alpha = 100

```python
rgd2 = Ridge(alpha=100)
rgd2.fit(X,y)
print(rgd2.coef_)
print(rgd2.intercept_)
```

```python
plt.plot(X,y,'b.')
plt.plot(X,lr.predict(X),color='red',label='alpha=0')
plt.plot(X,rgd1.predict(X),color='green',label='alpha=10')
plt.plot(X,rgd2.predict(X),color='orange',label='alpha=100')
plt.legend()
```

```python
class MeraRidgeRegression:
    def __init__(self, alpha=0.01):
        self.m=0
        self.b=0
        self.alpha=alpha
        
    def train(self,X_train,y_train):
        num =0
        den =0
        for i in range(X_train.shape[0]):
            
            num = num + (y_train[i]-y_train.mean()) * (X_train[i] - X_train.mean())
            den = den + ((X_train[i] - X_train.mean()) * (X_train[i] - X_train.mean())) 
        
        self.m = num / (den +self.alpha)
        self.b = y_train.mean() - (self.m * X_train.mean())
    
    def predict (self,X_test):
        return (self.X_test * self.m ) + self.b
        
```
